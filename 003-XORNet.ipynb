```python
# Import libraries
import torch
import torch.nn as nn
```

```python
nex = 100   # Number of examples input
X = torch.randint(0,2, size = (nex,2))  # input data randomely (binary)
y = torch.zeros(nex,1)  # init output
```

```python
# Calculate output to use in training
for i in range(nex):

    if (X[i,0] == 1 and X[i,1] == 0) or (X[i,0] == 0 and X[i,1] == 1):

        y[i,0] = 1
```

```python
# initializing model with sigmoed activation function and two neuron
AF = nn.Sigmoid()
model = nn.Sequential(nn.Linear(2, 2), AF,
                      nn.Linear(2, 1), AF)
```

```python
X = X.float()
y = y.float()
```

```python
# TRAIN
loss_fn = nn.MSELoss()  # init loss function to use
lr = 1e-1   # init learning rate
optimozer = torch.optim.Adam(model.parameters(), lr = lr)    # init our optimizer
model.train()   # start train
for t in range(3000):   # number of epochs
    
    y_pred = model(X)   # calculate predicted output
    loss = loss_fn(y_pred,y)    # calculate loss function 
    print(loss)
    optimozer.zero_grad()   # backward
    loss.backward() 
    optimozer.step() # optimize step of our weights

```

```python
s = torch.Tensor([0,1])
model(s)
```
